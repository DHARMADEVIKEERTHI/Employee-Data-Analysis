1. Missing Salary Imputation

```python
df['salary'] = df['salary'].fillna(df.groupby('department')['salary'].transform('median'))
```

3. Performance Score Conversion

```python
def convert_performance(score):
    try:
        return int(score)
    except (ValueError, TypeError):
        return -1

# Apply the conversion function to the 'performance_score' column
df['performance_score'] = df['performance_score'].apply(convert_performance)

# Check the unique values to verify
print("Unique values in performance_score after conversion:", df['performance_score'].unique())
```

4. Date Format Standardization

```python
join_year_list = []

for idx, row in df.iterrows():
    try:
        date = pd.to_datetime(row['join_date'], errors='raise')
        df.at[idx, 'join_date'] = date.strftime('%Y-%m-%d')
        join_year_list.append(date.year)
    except Exception:
        print(f"Error parsing date at index {idx}: {row['join_date']}")
        df.at[idx, 'join_date'] = None
        join_year_list.append(None)

df['join_year'] = join_year_list
```

5. Department Typo Correction

```python
department_corrections = {
    'HR': 'Human Resources',
    'Humman Resources': 'Human Resources',
    'human resourses': 'Human Resources',
    'Finanace': 'Finance',
    'FINANCE': 'Finance',
    'financee': 'Finance',
    'Engg': 'Engineering',
    'Eng': 'Engineering',
    'engineering': 'Engineering',
    'Marketingg': 'Marketing',
    'marketting': 'Marketing',
    'Sale': 'Sales',
    'Sales': 'Sales'
}

df['department'] = df['department'].replace(department_corrections)

df['department'] = df['department'].str.title()

print("Unique departments after correction:\n", df['department'].unique())
```

6. Salary Outlier Removal

```python
salary_mean = df['salary'].mean()
salary_std = df['salary'].std()

print(f"Mean Salary: {salary_mean}")
print(f"Standard Deviation of Salary: {salary_std}")

upper_limit = salary_mean + (3 * salary_std)
lower_limit = salary_mean - (3 * salary_std)

print(f"Lower = {lower_limit}, Upper = {upper_limit}")

df = df[(df['salary'] >= lower_limit) & (df['salary'] <= upper_limit)]

print(f"Data shape after removing outliers: {df.shape}")
```

7. Email Domain Extraction

```python
def extract_domain(email):
    try:
        if isinstance(email, str) and '@' in email:
            return email.split('@')[1]
        else:
            return None
    except Exception as e:
        print(f"Error processing email: {email}, Error: {e}")
        return None

df['email_domain'] = df['email'].apply(extract_domain)

print(df[['email', 'email_domain']].head())
```

8. Salary Correlation Calculation

```python
if df['join_year'].dropna().shape[0] > 1:
    correlation = df['salary'].corr(pd.to_numeric(df['join_year'], errors='coerce'))
    print(f"Pearson correlation (Salary vs Join Year): {correlation:.4f}\n")
else:
    print("Not enough valid data to calculate correlation.\n")
```

9. Average Salary Aggregation

```python
def assign_job_level(title):
    title = str(title).lower()
    if 'intern' in title:
        return 'Entry'
    elif 'junior' in title or 'associate' in title:
        return 'Junior'
    elif 'senior' in title or 'lead' in title:
        return 'Senior'
    elif 'manager' in title or 'head' in title:
        return 'Manager'
    elif 'director' in title or 'vp' in title:
        return 'Director'
    else:
        return 'Mid'

df['job_level'] = df['job_title'].apply(assign_job_level)
```

```python
pivot = pd.pivot_table(df, index='department', columns='job_level', values='salary', aggfunc='mean')
print("Pivot table (Average Salary by Department and Job Level):\n")
print(pivot)
```

10. Duplicate Row Removal

```python
duplicate_count_before = df.duplicated().sum()
print(duplicate_count_before)

df.drop_duplicates(inplace=True)

duplicate_count_after = df.duplicated().sum()
print(duplicate_count_after)
```

```python
df.isnull().sum() # Check Null Values
```

```python
null_percentage = (df.isnull().sum() / len(df)) * 100    # Check Percentage of Null Values
null_percentage
```

```python
df = df.dropna(subset=['first_name', 'last_name'], how='all') # Drop rows where both first_name and last_name are missing

df['first_name'] = df['first_name'].fillna('Unknown') # Replace missing last_name with 'Unknown'
df['last_name'] = df['last_name'].fillna('Unknown') # Replace missing first_name with 'Unknown'

df['email'] = df['email'].fillna('unknown@example.com') # Replace missing email with a default placeholder

# Extract domain from the email, handle missing/invalid emails
df['email_domain'] = df['email'].apply(
    lambda x: x.split('@')[1] if pd.notnull(x) and '@' in x else None
)
```

```python
df = df.dropna(subset=['join_date'])

def process_date(date):
    try:
        date_parsed = pd.to_datetime(date, errors='coerce') # Converts to datetime or NaT if invalid
        return date_parsed
    except:
        return pd.NaT # Fallback (though 'errors=coerce' handles most cases)

df['join_date'] = df['join_date'].apply(process_date) # Apply processed data
df = df.dropna(subset=['join_date']) # Drop null values
df['join_year'] = df['join_date'].dt.year # Extract new column year
```

```python
df['performance_score'] = pd.to_numeric(df['performance_score'], errors='coerce').fillna(-1).astype(int) # Convert 'performance_score' to numeric, replace errors
```
